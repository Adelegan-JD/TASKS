The following are the Mathematics test scores (out of 20) for a class of 15 students:
 5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20

**Solution**

numbers =  5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20

**Mean**

mean = sum of all values / number or count of all values

mean = (5 + 8 + 10 + 12 + 12 + 12 + 13 + 14 + 14 + 15 + 16 + 18 + 18 + 20 + 20) / 15

mean = 207 / 15

**mean = 13.8**


**Median**

median = middle number or mean of two middle numbers of ordered set of numbers

numbers =  5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20

median = **14**


**Mode**

mode = most frequently occuring number

numbers =  5, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20

**mode = 12**


**Replaced Value**:
If the lowest score (5) was wrongly recorded and should have been 10, how would the mean change? Would the median or mode change? Explain.

numbers =  10, 8, 10, 12, 12, 12, 13, 14, 14, 15, 16, 18, 18, 20, 20

A. The mean would increase from 13.8 t0 14.13 by 0.33

B. The median would remain the same as the replaced number does not fall in the middle of the total set of numbers, neither does it increase the total count of the set of numbers.

C. The mode also does not change because the replaced number does not occur more frequently than other numbers

2: Daily Commute Times
A group of 9 students reported the time (in minutes) it takes them to get to school each day:
15, 20, 20, 25, 25, 30, 60, 65, 120
Find the mean, median, and mode of the data.

Which measure (mean, median, or mode) do you think best represents the “typical” time it takes students to get to school? Why?
(Hint: Think about the effect of the student who takes 120 minutes.)


**Solution**

 number = 15, 20, 20, 25, 25, 30, 60, 65, 120


**Mean**

sum_of_numbers = 15 + 20 + 20 + 25 + 25 + 30 + 60 + 65 + 120 = 380
total_number_count = 9

mean = sum_of_numbers / total_number_count 

mean = 380 / 9

**mean = 42.2**



**Median**

To calculate the median, arrange the numbers in ascending or descending order

number = 15, 20, 20, 25, 25, 30, 60, 65, 120

**median = 25**



**Mode**
The modes are 20 and 25. Therefore, the data can be said to be bimodal



Which measure (mean, median, or mode) do you think best represents the “typical” time it takes students to get to school? Why?
(Hint: Think about the effect of the student who takes 120 minutes.)

I believe the median would be the most suitable measure of central tendecy to represent the typical time it takes for students to get to school. This is because the presence of an outlier (120) will affect the mean(because it is sensitive to outliers), causing a skewness in the data, and thus affecting the total result.
